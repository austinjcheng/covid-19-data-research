```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
CSV_FILE_PATH = './reduced_2020_Taxi_Trip_Data.csv'

df = pd.read_csv(CSV_FILE_PATH)
df.columns = ['month','trip_time','passenger','distance','rate','fare']

d_t = df.groupby('month')['trip_time'].mean().reset_index()
d_p = df.groupby('month')['passenger'].sum().reset_index()
d_m = df.groupby('month')['distance'].mean().reset_index()
d_f = df.groupby('month')['fare'].mean().reset_index()
d_r = df.groupby(['month','rate']).size().unstack()


cause = []
death = []
```

```python
with open('./cause_month','r') as f:
    for i in f.readlines():
        line = i.split()
        month = int(line[0].split('-')[1])
        x = int(line[1])
        cause.append([month, x])
list.sort(cause)
```

```python
with open('./death_month','r') as f:
    for i in f.readlines():
        line = i.split()
        month = int(line[0].split('-')[1])
        x = int(line[1])
        death.append([month, x])
list.sort(death)
```

```python
cause = pd.DataFrame(cause, columns=["month", "cases"])
death = pd.DataFrame(death, columns=["month", "deaths"])
combined_covid = pd.merge(cause,death)
```

```python
fig, ax = plt.subplots() 
ax.plot(combined_covid['month'],combined_covid['cases'], label='cases')
ax.plot(combined_covid['month'],combined_covid['deaths'], label='deaths', color='red')
ax.ticklabel_format(style='plain')
ax.legend()

plt.xlabel("Month")
plt.ylabel("# of People")
plt.xticks(list(range(1,13)))
plt.show()
```
